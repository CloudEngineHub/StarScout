```python
import random
from pymongo import MongoClient
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from scripts import MONGO_URL, GITHUB_TOKENS
from scripts.analysis import data, plot

from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
plot.init_matplotlib()
```

```python
client = MongoClient(MONGO_URL)
db = client['fake_stars']
collection = db['actors']
sample_size = 10000
headers = {
    'Authorization': f'token {GITHUB_TOKENS[0]}',
    'Accept': 'application/vnd.github.v3+json',
    'User-Agent': 'YourAppName'  # Replace with your app name or leave as is
}
# Sample 10,000 actors randomly
sampled_actors_cursor = collection.aggregate([{'$sample': {'size': sample_size}}])

invalid = 0
# Process each actor
for actor in sampled_actors_cursor:
    actor_id = actor.get('actor')
    info = actor.get('info')
    
    if info is None:
        invalid += 1
    else:
        # Access the 'orgs' field within 'info'
        orgs = info.get('organizations_url')
        if not orgs:
            invalid += 1
        else:
            response = requests.get(orgs,headers=headers)
            if len(response) == 0:
                invalid += 1
```
